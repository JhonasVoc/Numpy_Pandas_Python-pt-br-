## Preenchendo dados faltantes fillna()

Podemos alterar elementos com o replace(), mas no nossa casoq ueremos alterar todos os dados faltantes

```python
data = {'Cidades':['Porto Alegre','São Carlos',np.nan],
        'Estado':['Paraná',np.nan,'São Paulos'],
        'Países':['Brasil',np.nan,np.nan]}
data = pd.DataFrame(data)
data['Cidades'] = data['Cidades'].replace({np.nan:'Santa Maria'})
data
```

<font color = blue> Usando o fillna(), podemos substituir todos os dados faltantes na nossa variável

```python
data = {'Cidades':['Porto Alegre','São Carlos',np.nan],
        'Estado':['Paraná',np.nan,'São Paulos'],
        'Países':['Brasil',np.nan,np.nan]}
data = pd.DataFrame(data)
data = data.fillna(value ='Dados Ausentes')
data
```

Podemos usar de toda e qualquer operação matemática sobre os valores nummméricos

```python
data = {'1 Semestre':[8.2,8.0,np.nan],
        '2 Semestre':[7.7,np.nan,8.0],
        'Recuperação':[7.9,np.nan,np.nan]}
data = pd.DataFrame(data)
data
```

```python
#outro método é usar o fillna com o ffill, o qual substitui os valores ausentes pelos dados antecessores
data = {'1 Semestre':[8.2,8.0,np.nan],
        '2 Semestre':[7.7,np.nan,8.0],
        'Recuperação':[7.9,np.nan,np.nan]}
data = pd.DataFrame(data)

data = data.fillna(method = 'ffill')
data
```

# Processando em paralelo de um um DataFrame

    Quando estamos trabalhando com grande volume de dados, eficiência é um assunto muito importante, sendo assim,
    paralelismo é uma técnica muito útil.
    Basicamente, consiste em dividir deteminada tarefa ou função, aumentando a performance e velocidade.
    

```python
from multiprocessing import Pool
```

```python
#Pool permite definir a execução de uma função sobre nossos dado em paralelistmo
```

```python
base  = "new_york_hotels.csv"
```

```python
df = pd.read_csv(base, encoding = "ISO-8859-1") # ecoding é usado apra definirmos os dados de maneira compativel, se não ele é lido como script
df.head(10)
```

```python
# Vamos  calcula a distância entre os dois pontos, usando os dados de latitude e longitude via fórmula de Haversine
```

```python
def haversine(df):
    lat1, lon1, lat2, lon2 = 40.671,-73.985, df['latitude'].values, df['longitude'].values
    MILES  = 3959
    lat1, lon1, lat2, lon2 = map (np.deg2rad,[lat1,lon1,lat2,lon2])
    dlat = lat2  - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) *np.cos(lat2) * np.sin(dlon/2)**2
    c = 2* np.arcsin(np.sqrt(a))
    total_miles = MILES * c
    df['distance'] = total_miles
    return df
```

```python
%%timeit
distance_df = haversine(df)
```

```python
def parallelize_dataframe(df,func,n_cores = 8):
    df_split = np.array_split(df,n_cores)
    pool = Pool(n_cores)
    df = pd.concat(pool.map(func,df_split))
    pool.close()
    pool.join()
    return df
```

```python
%%timeit
distance_df= parallelize_dataframe(df,haversine)
```
